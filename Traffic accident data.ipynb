```python
# Handle Missing Values
# For df_accidents, impute missing numerical values with the mean and categorical values with the mode.
# Remove rows with missing values in crucial columns (e.g., Accident_Severity)
for col in df_accidents.select_dtypes(include=['number']):
    df_accidents[col] = df_accidents[col].fillna(df_accidents[col].mean())
for col in df_accidents.select_dtypes(include=['object']):
    df_accidents[col] = df_accidents[col].fillna(df_accidents[col].mode()[0])

# Remove rows with missing values in 'Accident_Severity' as it's crucial
df_accidents.dropna(subset=['Accident_Severity'], inplace=True)

# For df_casualties and df_vehicles, impute or remove rows with missing values based on the column importance
for col in df_casualties.select_dtypes(include=['number']):
    df_casualties[col] = df_casualties[col].fillna(df_casualties[col].mean())
for col in df_casualties.select_dtypes(include=['object']):
    df_casualties[col] = df_casualties[col].fillna(df_casualties[col].mode()[0])

for col in df_vehicles.select_dtypes(include=['number']):
    df_vehicles[col] = df_vehicles[col].fillna(df_vehicles[col].mean())
for col in df_vehicles.select_dtypes(include=['object']):
    df_vehicles[col] = df_vehicles[col].fillna(df_vehicles[col].mode()[0])

# Remove Duplicates
duplicates_accidents = df_accidents.duplicated().sum()
df_accidents.drop_duplicates(inplace=True)
duplicates_casualties = df_casualties.duplicated().sum()
df_casualties.drop_duplicates(inplace=True)
duplicates_vehicles = df_vehicles.duplicated().sum()
df_vehicles.drop_duplicates(inplace=True)

print(f"Removed {duplicates_accidents} duplicate rows from df_accidents")
print(f"Removed {duplicates_casualties} duplicate rows from df_casualties")
print(f"Removed {duplicates_vehicles} duplicate rows from df_vehicles")

# Data Type Conversion
df_accidents['Date'] = pd.to_datetime(df_accidents['Date'], format='%d-%m-%Y', errors='coerce')
# Handle errors in Date conversion

# Address Data Inconsistencies
# Ensure 'Accident_Index' consistency across dataframes
# Convert 'Accident_Index' to the same data type if necessary
df_accidents['Accident_Index'] = df_accidents['Accident_Index'].astype(int)
df_casualties['Accident_Index'] = df_casualties['Accident_Index'].astype(int)
df_vehicles['Accident_Index'] = df_vehicles['Accident_Index'].astype(int)

# Display the first few rows of each DataFrame after cleaning
display(df_accidents.head())
display(df_casualties.head())
display(df_vehicles.head())
```

```python
# Merge DataFrames
df_merged = pd.merge(df_accidents, df_casualties, on='Accident_Index', how='left')
df_merged = pd.merge(df_merged, df_vehicles, on='Accident_Index', how='left')

# Create 'Time_of_Day' feature
def categorize_time(time_str):
    try:
        hour = int(time_str.split(':')[0])
        if 6 <= hour < 12:
            return 'Morning'
        elif 12 <= hour < 17:
            return 'Afternoon'
        elif 17 <= hour < 20:
            return 'Evening'
        else:
            return 'Night'
    except (ValueError, AttributeError, IndexError):
        return 'Unknown'  # Handle cases with missing or invalid time values
    
df_merged['Time_of_Day'] = df_merged['Time'].apply(categorize_time)

# Create 'Combined_Conditions' feature
df_merged['Combined_Conditions'] = df_merged['Weather_Conditions'].astype(str) + ', ' + df_merged['Road_Surface_Conditions'].astype(str)

# Example of an additional feature combination (Time_of_Day and Weather_Conditions)
df_merged['Time_Weather'] = df_merged['Time_of_Day'] + ', ' + df_merged['Weather_Conditions'].astype(str)

# Display the head of the merged DataFrame
display(df_merged.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Accident Frequency Analysis
time_of_day_counts = df_merged['Time_of_Day'].value_counts()
weather_counts = df_merged['Weather_Conditions'].value_counts()
road_surface_counts = df_merged['Road_Surface_Conditions'].value_counts()

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.barplot(x=time_of_day_counts.index, y=time_of_day_counts.values)
plt.title('Accident Frequency by Time of Day')
plt.subplot(1, 3, 2)
sns.barplot(x=weather_counts.index, y=weather_counts.values)
plt.title('Accident Frequency by Weather')
plt.subplot(1, 3, 3)
sns.barplot(x=road_surface_counts.index, y=road_surface_counts.values)
plt.title('Accident Frequency by Road Surface')
plt.tight_layout()
plt.show()

# Accident Severity Analysis
severity_by_time = df_merged.groupby('Time_of_Day')['Accident_Severity'].value_counts(normalize=True).unstack()
severity_by_weather = df_merged.groupby('Weather_Conditions')['Accident_Severity'].value_counts(normalize=True).unstack()
severity_by_road = df_merged.groupby('Road_Surface_Conditions')['Accident_Severity'].value_counts(normalize=True).unstack()

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
severity_by_time.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Accident Severity by Time of Day')
plt.subplot(1, 3, 2)
severity_by_weather.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Accident Severity by Weather')
plt.subplot(1, 3, 3)
severity_by_road.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Accident Severity by Road Surface')
plt.tight_layout()
plt.show()


# Combined Condition Analysis
combined_conditions_counts = df_merged['Combined_Conditions'].value_counts()
severity_by_combined = df_merged.groupby('Combined_Conditions')['Accident_Severity'].value_counts(normalize=True).unstack()

plt.figure(figsize=(10, 6))
sns.heatmap(severity_by_combined, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title('Accident Severity by Combined Weather and Road Conditions')
plt.show()

# Time and Weather Combined Analysis
time_weather_counts = df_merged['Time_Weather'].value_counts()
severity_by_time_weather = df_merged.groupby('Time_Weather')['Accident_Severity'].value_counts(normalize=True).unstack()

plt.figure(figsize=(12, 6))
severity_by_time_weather.plot(kind='bar', stacked=True)
plt.title('Accident Severity by Time of Day and Weather')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
